# Measuring impact of model assumptions in binary black hole simuations  - 

## Binary Black Holes - 

A system of two black holes gravitationally bound to each other loose energy with time and eventually results in merger forming a more massive black hole. As these black holes evolve, they perturb the structure of spacetime creating ripples which travel as gravitational waves. These waves carry the information about the source binary, and their detection allows us to learn about these systems even from a distance of millions of light years. Observations of such binaries provide us a new way to learn more about the origin and evolution of universe, provide new ways to gauge into its hidden corners, explore the possibilities of dark matter and dark energy and test fundamental theories of physics. 



Detection of such systems relies on our ability to accurately model their dynamics and their gravitational wave signatures. For this purpose, we perform numerical simulations to solve a set of coupled partial differential equations (Einstein's equations) in full 3-d space. To solve a partial differential equation requires initial conditions (state of system at initial time t=0) and boundary conditions (how solution behaves near boundaries) which can affect the accuracy of numerical solutions. __Here, we analyse the effects of initial data on binary black hole simulation.__ 



## Model Assumption and Side Effects

Finding the optimal initial data is a complex problem  and generally carries certain assumption about the surface geometry. Such assumptions can have additional side effects. 


__Binary Black Holes Initial Data__ - A black hole is a geometrical object which can be completely characterized by its mass, spin and momenta (with respect to another observer at rest). A system of binary black holes consists of two black holes of initial masses $m_1$, $m_2$, and initial spins $\vec{a}_1$, $\vec{a}_2$ separated by a distance $D$. At initial time $t=0$ the black holes have some momenta $\vec{p}_1$ and $\vec{p}_2$. Using these set of parameters, we can uniquely define the 3-d surface metric and curvature at initial time where black holes reside. 

![Binary Black Holes](bbh.jpg)

To define the geometry of spacetime, we assume conformal space to be flat which is not completely true. This leads to artificial burst of radiation in the beginning of the simulation as system relaxes into its true state. Such burst of energy contaminates the signal with noise (check the figure below). Further, some of this radiated energy is absorbed by the black holes thus altering the properties of the binary system (such as black hole mass and spin). This change is equivalent to simulating black hole binary system with different initial conditions in absence of such radiation. As the final configuration does not represent the desired initial conditions, this is a clear side effect.

## Initial Hypothesis and Prediction - 

__Hypothesis__ - Analyzing the data, we see that the change in the properties of two black holes occur independent of each other. In other words, we assume that the change in black hole mass and spin only depends on the infall of radiation into black hole's surface and is not affected by the presence of nearby black hole in any way. Hence, we can approximate each black hole to be in isolated state.

__Prediction__ - Model of isolated black hole should also predict similar changes in mass and spins as seen for black holes in binary. 

## Testing - 

To test this hypothesis, we perform around 120 simulations of single black hole for varying spin and initial speed using the same data method. After some initial analysis, we compute the changes in mass and spin of black hole for each case. Using a fifth order polynomial fitting, we are able to show that change in mass and spin show polynomial behavior as a function of initial spin and initial speed. 

Figure below shows that mass changes and spin changes as a function of initial speed and initial spin for a single black hole

## Model Correction - 

The modifications in source properties are undesirable and affects the validity of the model. In particular, it would be preferred if we could specify the initial state of the system in its relaxed state after passing the junk radiation. This would account for mass and spin changes from artificial radiation and ensure we are modeling the true state. 

Here, we devise a new method based on above methods. We start with desired configuration of the binary parameters described by masses $M_1, M_2$ and spin magnitudes $\vec{a}_1, \vec{a}_2$. Then, at $t=0$, the configuration of the binary can be determined by solving the following equations - 

$$ M_i = [1 + \delta \bar{M}_i \,(\bar{a}_i, v_i)]\,\bar{M}_i$$
$$ a_i = [1 + \delta \bar{a}_i \,(\bar{a}_i, v_i)]\,\bar{a}_i$$

Here, $\bar{M}_i$ and $\bar{a}_i$ are the masses and spin magnitudes of $i^{th}$ black hole at initial time. The above equations can be solved using Newton's method. Below, we show the gravitational wave signatures from the original and the corrected initial data for spin magnitudes of 0.6 and 0.8 using this method. As can be seen, the waveform signatures are clearly different. 

```python
# GW signatures of original and corrected simulation

os.chdir('/localdata2/bkhamesra3/simulations/BBH/Comet/JunkRadiation/BBH/GenerateStrain/WFSummaryDirectory/')

#Read the data

t1, re1, im1 = np.loadtxt('./BBH_JK_q1_a0.6_M150-all/Ylm_WEYLSCAL4::Psi4r_l2_m2_r75.00.asc', unpack=True, usecols=(0,1,2))
t2, re2, im2 = np.loadtxt('./BBH_JK_q1_a0.6_M150_IDUpdated_Fits-all/Ylm_WEYLSCAL4::Psi4r_l2_m2_r75.00.asc', unpack=True, usecols=(0,1,2))

t3, re3, im3 = np.loadtxt('./BBH_JK_q1_a0.8_M150-all/Ylm_WEYLSCAL4::Psi4r_l2_m2_r75.00.asc', unpack=True, usecols=(0,1,2))
t4, re4, im4 = np.loadtxt('./BBH_JK_q1_a0.8_M150_IDUpdated_Fits-all/Ylm_WEYLSCAL4::Psi4r_l2_m2_r75.00.asc', unpack=True, usecols=(0,1,2))


#Plot the results
fig, (ax1,ax2)=plt.subplots(2,1,figsize=(10,10), sharex=True)
plt.subplots_adjust(wspace=0, hspace=0.05)
ax1.plot(t1, re1, ls='--',label='Original Simulation')
ax1.plot(t2, re2, label='Corrected Simulation')
#ax1.set_xlabel('Time')
ax1.set_ylabel(r'$Re(\Psi_4)$', fontsize=20)
ax1.set_xlim(0,700)
ax1.ticklabel_format(style='sci', scilimits=(0,3))
ax1.tick_params(axis='both',labelsize=18)

ax1.legend()


ax2.plot(t3, re3, ls='--',label='Original Simulation')
ax2.plot(t4, re4, label='Corrected Simulation')
ax2.set_xlabel('Time (M)', fontsize=20)
ax2.set_ylabel(r'$Re(\Psi_4)$', fontsize=20)
ax2.set_xlim(0,700)
ax2.ticklabel_format(style='sci', scilimits=(0,3))
ax2.tick_params(axis='both',labelsize=18)


ax2.legend()
plt.tight_layout()
plt.show()
plt.close()


```
