```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl

#Basic plot settings
mpl.rc('lines', linewidth=2, color='r')
mpl.rc('font', size=16)
mpl.rc('axes', labelsize=16, grid=True)
mpl.rc('xtick', labelsize=14)
mpl.rc('ytick', labelsize=14)
mpl.rc('legend', fontsize=14)
```

```python
#Creating a sine and cosine wave - 
theta = np.linspace(0,360, 1000)
x = theta/180.0*np.pi
sinx = np.sin(x)
cosx = np.cos(x)

#Create figure
plt.figure(figsize=(12,6))

#Line plot
plt.plot(theta, sinx, label='Sine') 
plt.plot(theta, cosx, label='Cosine') 
#Label axes
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f(\theta)$')
plt.legend()
plt.show()
plt.close()
```
